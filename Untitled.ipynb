## Read in covid data
The Office of National Statistics (ONS) has published the number of deaths due to covid and not-covid in each middle-layer super output area (MSOA) between March and June 2020.

[*Deaths by MSOA (ONS)*](https://www.ons.gov.uk/peoplepopulationandcommunity/birthsdeathsandmarriages/deaths/datasets/deathsinvolvingcovid19bylocalareaanddeprivation)

```python
# pandas let's you interact with data in a much nicer way that the lists and arrays it's built on.
import pandas as pd 

# read in the data from "Table 5" in the covid spreadsheet
covid_deaths = pd.read_excel(io='referencetablesworkbook2.xlsx', # in/out file
                      sheet_name='Table 5',                      # data sheet
                      header=12,                                 # skip 12 lines
                      names=['msoa_code',                        # define column names
                             'msoa_name',
                             'common_name',
                             'march_covid',
                             'april_covid',
                             'may_covid',
                             'june_covid',
                             'sum_covid',
                             'march_other',
                             'april_other',
                             'may_other',
                             'june_other',
                             'sum_other',
                             'march_all',
                             'april_all',
                             'may_all',
                             'june_all',
                             'sum_all',
                            ],
                      usecols="A:C,E:I,K:O,Q:U"                # use these columns from the spreadsheet
                     )

# show first 5 lines of `covid_deaths` data
covid_deaths.head(5)
```

## Aggregate up to electoral ward level
The Office of National Statistics provides lookup tables to aggregate postcodes up to higher-level geographies using a "best fit" approach where geographies aren't perrfectly aligned. There are something like 2 million postcodes in the UK, so it might take a minute to process. You won't even be able to open the file in excel.   
We're using the most recent version here.
[*Postcode lookup tables (ONS)*](https://geoportal.statistics.gov.uk/datasets/national-statistics-postcode-lookup-may-2020)

## Get the lookup table...

```python
# read in the file of all postcodes
msoas = pd.read_csv('./NSPL_FEB_2020_UK/Data/NSPL_FEB_2020_UK.csv')

# get rid of everything other than MSOA codes and their corresponding 'best fit' parliamentary constituencies
msoas = msoas[['msoa11','pcon']]

# each line was a unique postcode, so with multiple postcodes to each MSOA, and multiple MSOAs to each constituency,
# most of these rows are redundant. Drop them like it's hot.
msoas.drop_duplicates(inplace=True) # inplace=True overwrites the old table with the new (smaller) one.

# rename the columns as something readable
msoas.rename(columns={'msoa11': 'msoa_code',
                      'pcon': 'constituency_code'
                     },
            inplace=True)

# show the first few rows
msoas.head(10)

# A warning message might pop up. It's fine.
```

## Get the names for parliamentary constituencies and add them to the lookup table
The lookup table isn't very human-readable. In particular, we're missing the names for the constituencies, so we'll grab them from the supplementary data that came with the lookup tabel.   
The MSOA names are already in the covid dataframe, so we don't need to get them too.

### Get the data...

```python
# read in the table with the keys
constituancy_key = pd.read_csv(
    './NSPL_FEB_2020_UK/Documents/Westminster Parliamentary Constituency names and codes UK as at 12_14.csv')

constituancy_key.head()
```

### ...and make the join

```python
# join onto the msoas table. When A.join(B)-ing two dataframes, the index in frame B must correspond
# to one of the columns in frame A. Values in B.index and A.column are compared, and rows are matched
# to add values in B to the right records in A. This is very similar to joins in SQL. 
# on=A.column specifies the column in A on which to attempt the join.

msoas_with_words = msoas.join(constituancy_key.set_index('PCON14CD'), on = 'constituency_code')

# remane 'PCON14NM' because it's a stupid name and we're going to be carrying it forward
msoas_with_words.rename(columns={'PCON14NM': 'constituency_name'}, inplace=True)

# show the first 5 rows
msoas_with_words.head()
```

## Join the lookup table to the covid data
We now have a many-to-one mapping between MSOAs and parliamentary constituencies. We can therefore use the MSOA codes to join this map to the covid data. The end result will be the original covid dataframe with two new columns: `constituency_code` and `constituency_name`.

```python
# covid_tba (to be aggregated)--a temporary table to break things down into nicer steps
# how='left' means "if this value exists in covid_deaths' msoa_code column, then join the 
# corresponding value from msoas_with_words to that row, otherwise don't add this value."
covid_tba = covid_deaths.join(msoas_with_words.set_index('msoa_code'), on='msoa_code', how='left')
covid_tba.head()

# Success!
```

## Aggregate up to constituency level
For each parliametnary constituency, we need to 
 1. isolate only the rows of covid_tba which correspond to msoas which fall within that constituency
 2. seperately add up all the different statistics
 3. add that constituency and its values to a new dataframe
 
We can do this by creating a `groupby` object and specifying the aggregation function as a summation.

First, create the `groupby object`...

```python
# create the groupby object
groupby_object = covid_tba.groupby('constituency_name')

# groupby objects don't really look like anything until you give them an aggregation function:
print("Our groupby object:", groupby_object)
```

...and then pass 'sum' as the aggregation argument

```python
covid_constituencies = groupby_object.sum()
covid_constituencies
```

Notice that we only have 573 rows rather than the expected 650 parliamentary seats. That's because our covid data is only for England (533) seats) and Wales (40 seats).

Great! We have numbers for the covid and non-covid deaths in England and Wales between March and June. Now for the election results.

## Get the election data
Election results are available from the House of Commons Library as a .csv file. It includes the name of the elected MP, the party, whether the seat changed parties, and the vote breakdown. For our purposes, let's begin by looking at the winning party.

[*Election Results: House of Commons Library*](https://commonslibrary.parliament.uk/research-briefings/cbp-8749/)

```python
election_result = pd.read_csv('./HoC-GE2019-results-by-constituency-csv.csv')
election_result = election_result[['constituency_name', 'first_party']]
election_result
```

Now we can join the election results to the covid data by constituency:

```python
# left join since our covid data is only for England ans Wales
political_covid = covid_constituencies.join(election_result.set_index('constituency_name'), how='left')
political_covid
```

This dataframe is getting a bit unwieldy, so I'm going to make a copy and drop everything apart from the total covid deaths and the winning party. We can always go back to the full dataset later. I'll also give them better column names.

```python
covid_deaths_only = political_covid[['sum_covid', 'first_party']].copy()
covid_deaths_only.rename(columns={'sum_covid': 'covid_deaths', 'first_party': 'party'}, inplace=True)
covid_deaths_only
```

## Making a map
To make a map, we need shapefiles which describe the boundaries of the constituencies. These can be downloaded from geoportal.statistics.gov.uk

[*Shapefiles*](https://geoportal.statistics.gov.uk/datasets/westminster-parliamentary-constituencies-december-2017-uk-bgc)

```python
import geopandas as gpd         # pandas for maps
import matplotlib.pyplot as plt # for plotting
%matplotlib inline              

# the file path is long, so place it in the variable `my_file` to make code more readable
my_file = 'Westminster_Parliamentary_Constituencies__December_2017__UK_BGC_V2'
#           Westminster_Parliamentary_Constituencies__December_2017__UK_BGC_V2-shp
constituency_map = gpd.read_file(my_file + "-shp/" + my_file + ".shp")
constituency_map.plot()
plt.show()
constituency_map.head()
```

## Add the data to the map
Since the map data is in a geopandas frame, we can use joins just like we did above. We just need to make sure that the **geo**pandas frame is on the left.

```python
# no need to .set_index(), since covid_deaths_only is already indexed on constituency
map_with_data = constituency_map.join(covid_deaths_only, on='PCON17NM', how='right')
map_with_data.head()
```

## Actually plot a map!


```python
fig, ax = plt.subplots(figsize=(12,24))

map_with_data.plot(ax=ax, column='covid_deaths', cmap='Blues')
plt.show()
```

```python
map_with_data.to_file("map_data.gpkg", layer = "covid_deaths", driver="GPKG")
map_with_data.to_file("map_data.gpkg", layer = "party", driver="GPKG")
#msoas_with_data.to_file("C:/Users/Thomas Walsh/Desktop/IMD_MSOAs.gpkg", layer='IMD_decile', driver="GPKG")
```
